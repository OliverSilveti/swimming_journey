## Filter for JSON files

### Subtask:
Identify and filter the list to include only JSON files.

**Reasoning**:
Iterate through the `file_list` and create a new list containing only the files that end with the `.json` extension.

```python
json_files = [f for f in file_list if f.endswith('.json')]
print(json_files)
```

```python
import json

folder_path = '/content/drive/MyDrive/Suunto'

workout_data = []

for file_name in json_files:
  file_path = os.path.join(folder_path, file_name)
  with open(file_path, 'r') as f:
    data = json.load(f)
    workout_data.append(data)

print(f"Successfully read {len(workout_data)} JSON files.")
```

```python
# Assuming json_files contains the list of JSON filenames
workout_types = {}

for file_name in json_files:
    file_name_lower = file_name.lower()
    if 'openwater' in file_name_lower:
        workout_types[file_name] = 'Open Water'
    elif 'swimming' in file_name_lower:
        workout_types[file_name] = 'Pool'
    else:
        workout_types[file_name] = 'Unknown'

print("Identified workout types based on filenames:")
for file_name, workout_type in workout_types.items():
    print(f"- {file_name}: {workout_type}")

# Optional: Store the workout type in the workout_data structure if needed later
# for workout_data_entry in workout_data:
#    file_name = workout_data_entry.get('file_name') # Assuming we add file_name to workout_data earlier
#    if file_name:
#        workout_data_entry['workout_type'] = workout_types.get(file_name, 'Unknown')
```

## Extract Data and Calculate Pace per Workout

### Subtask:
Extract distance and calculate moving time (or relevant duration) and pace for each workout based on its type.

**Reasoning**:
Iterate through the workout data, use the identified workout type to determine how to extract distance and calculate moving time (or a relevant duration like active duration from header or interval duration), and then calculate the pace.

```python
import pandas as pd

workout_summary_list = []

for workout_index, workout in enumerate(workout_data):
    file_name = json_files[workout_index] # Assuming json_files list order matches workout_data
    workout_type = workout_types.get(file_name, 'Unknown')

    distance_meters = None
    duration_seconds = None
    moving_time_seconds = None # We will use this to store the relevant duration for pace calculation

    pace_minutes_per_100m = None
    # Removing heart rate and stroke rate for now based on user request
    # average_heart_rate = None
    # average_stroke_rate = None


    if 'DeviceLog' in workout and 'Header' in workout['DeviceLog']:
        header = workout['DeviceLog']['Header']
        distance_meters = header.get('Distance')
        total_duration_seconds = header.get('Duration')
        pause_duration_seconds = header.get('PauseDuration', 0)

        # Use header data for open water workouts
        if workout_type == 'Open Water':
            duration_seconds = total_duration_seconds
            moving_time_seconds = total_duration_seconds - pause_duration_seconds # Use active duration as moving time proxy
            print(f"\nProcessing Open Water Workout ({file_name}):")
            print(f"  Distance: {distance_meters} m, Total Duration: {total_duration_seconds:.2f} s, Pause Duration: {pause_duration_seconds:.2f} s")

        # For Pool workouts, let's try using the total duration/distance from 'Interval' windows if available,
        # as the user's expected pace seemed closer to the interval pace.
        # If no intervals, fall back to header total duration - pause, or just total duration.
        elif workout_type == 'Pool':
             print(f"\nProcessing Pool Workout ({file_name}):")
             interval_duration = 0
             interval_distance = 0
             if 'Windows' in workout['DeviceLog']:
                windows_data = workout['DeviceLog']['Windows']
                for window_entry in windows_data:
                    window = window_entry.get('Window')
                    if window and window.get('Type') == 'Interval':
                        interval_duration += window.get('Duration', 0)
                        interval_distance += window.get('Distance', 0)

             if interval_distance > 0:
                 moving_time_seconds = interval_duration
                 distance_meters = interval_distance # Use interval distance for pace calculation
                 duration_seconds = total_duration_seconds # Keep total duration for context
                 print(f"  Using Interval data - Distance: {interval_distance} m, Duration: {interval_duration:.2f} s")
             else:
                 # Fallback if no interval data
                 duration_seconds = total_duration_seconds
                 moving_time_seconds = total_duration_seconds - pause_duration_seconds # Fallback to active duration
                 print(f"  No Interval data found, falling back to Header data - Distance: {distance_meters} m, Active Duration: {moving_time_seconds:.2f} s")


    # Calculate Pace if we have moving time and distance
    if moving_time_seconds is not None and distance_meters is not None and distance_meters > 0 and moving_time_seconds > 0:
        pace_seconds_per_meter = moving_time_seconds / distance_meters
        pace_seconds_per_100_meters = pace_seconds_per_meter * 100
        minutes = int(pace_seconds_per_100_meters // 60)
        seconds = round(pace_seconds_per_100_meters % 60)
        pace_minutes_per_100m = f"{minutes}:{seconds:02d}"
        print(f"  Calculated Pace: {pace_minutes_per_100m} per 100m")
    else:
        print("  Could not calculate pace.")

    # Extract and calculate average Heart Rate and Stroke Rate from Samples
    # if 'DeviceLog' in workout and 'Samples' in workout['DeviceLog']:
    #     samples = workout['DeviceLog']['Samples']
    #     heart_rates = []
    #     stroke_rates = []

    #     for sample in samples:
    #         if 'HR' in sample and sample['HR'] is not None:
    #             heart_rates.append(sample['HR'])
    #         if 'StrokeRate' in sample and sample['StrokeRate'] is not None:
    #             stroke_rates.append(sample['StrokeRate'])

    #     average_heart_rate = sum(heart_rates) / len(heart_rates) if heart_rates else None
    #     average_stroke_rate = sum(stroke_rates) / len(stroke_rates) if stroke_rates else None

    #     if average_heart_rate is not None:
    #         print(f"  Average Heart Rate: {average_heart_rate:.2f}")
    #     if average_stroke_rate is not None:
    #         print(f"  Average Stroke Rate: {average_stroke_rate:.2f}")


    # Append data to the list
    workout_summary_list.append({
        'FileName': file_name,
        'WorkoutType': workout_type,
        'Distance_meters': distance_meters,
        'TotalDuration_seconds': duration_seconds, # Store total duration for context
        'MovingTime_seconds': moving_time_seconds, # Store the duration used for pace calculation
        'AveragePace_min_per_100m': pace_minutes_per_100m,
        # 'AverageHeartRate': average_heart_rate, # Removing heart rate and stroke rate for now
        # 'AverageStrokeRate': average_stroke_rate
    })

# Create DataFrame
workout_summary_df = pd.DataFrame(workout_summary_list)

print("\nWorkout Summary DataFrame (Heart Rate and Stroke Rate excluded):")
display(workout_summary_df)
```

## Prepare data for regression

### Subtask:
Ensure the data is in a suitable format for linear regression, converting the 'DateTime' to a numerical representation (e.g., timestamps or days since the first workout).


**Reasoning**:
Convert 'DateTime' to a numerical representation (days since the start) and handle potential missing values in columns for regression.



```python
# Convert 'DateTime' to days since the first workout
workout_summary_df['Days_Since_Start'] = (workout_summary_df['DateTime'] - workout_summary_df['DateTime'].min()).dt.days

# Handle potential missing values in the columns used for regression.
# We will use the smoothed columns which already have interpolated values.
# However, we still need to handle any potential NaNs that might remain at the edges
# after smoothing if the window size is larger than the number of initial rows.
# A simple forward fill followed by a backward fill can handle these edge cases.
workout_summary_df['Smoothed_Distance_meters'].fillna(method='ffill', inplace=True)
workout_summary_df['Smoothed_Distance_meters'].fillna(method='bfill', inplace=True)

workout_summary_df['Smoothed_MovingTime_seconds'].fillna(method='ffill', inplace=True)
workout_summary_df['Smoothed_MovingTime_seconds'].fillna(method='bfill', inplace=True)

workout_summary_df['Smoothed_Pace_seconds_per_100m'].fillna(method='ffill', inplace=True)
workout_summary_df['Smoothed_Pace_seconds_per_100m'].fillna(method='bfill', inplace=True)

# Display the DataFrame with the new 'Days_Since_Start' column and check for NaNs in smoothed columns
display(workout_summary_df[['FileName', 'DateTime', 'Days_Since_Start',
                            'Smoothed_Distance_meters', 'Smoothed_MovingTime_seconds',
                            'Smoothed_Pace_seconds_per_100m']].head())
display(workout_summary_df[['Smoothed_Distance_meters', 'Smoothed_MovingTime_seconds', 'Smoothed_Pace_seconds_per_100m']].isnull().sum())
```

## Perform linear regression

### Subtask:
For each metric (distance, moving time, pace), fit a linear regression model with time as the independent variable.


**Reasoning**:
Import LinearRegression and fit models for each smoothed metric using 'Days_Since_Start' as the independent variable.



```python
from sklearn.linear_model import LinearRegression

# Fit linear regression model for Smoothed_Distance_meters
distance_model = LinearRegression()
X = workout_summary_df['Days_Since_Start'].values.reshape(-1, 1)
y = workout_summary_df['Smoothed_Distance_meters']
distance_model.fit(X, y)

# Fit linear regression model for Smoothed_MovingTime_seconds
moving_time_model = LinearRegression()
X = workout_summary_df['Days_Since_Start'].values.reshape(-1, 1)
y = workout_summary_df['Smoothed_MovingTime_seconds']
moving_time_model.fit(X, y)

# Fit linear regression model for Smoothed_Pace_seconds_per_100m
pace_model = LinearRegression()
X = workout_summary_df['Days_Since_Start'].values.reshape(-1, 1)
y = workout_summary_df['Smoothed_Pace_seconds_per_100m']
pace_model.fit(X, y)

print("Linear regression models fitted for smoothed distance, moving time, and pace.")
```

## Predict trend line

### Subtask:
Predict trend line


**Reasoning**:
Generate a range of 'Days_Since_Start' values for prediction and use the fitted models to predict the trend lines.



```python
import numpy as np

# Create a range of 'Days_Since_Start' values for prediction
# Start from the minimum Days_Since_Start to the maximum, creating a smooth range
X_predict = np.linspace(workout_summary_df['Days_Since_Start'].min(),
                        workout_summary_df['Days_Since_Start'].max(),
                        100).reshape(-1, 1) # Create 100 points for a smooth line

# Predict the smoothed metric values using the fitted models
workout_summary_df['Distance_Trend'] = distance_model.predict(workout_summary_df['Days_Since_Start'].values.reshape(-1, 1))
workout_summary_df['MovingTime_Trend'] = moving_time_model.predict(workout_summary_df['Days_Since_Start'].values.reshape(-1, 1))
workout_summary_df['Pace_Trend'] = pace_model.predict(workout_summary_df['Days_Since_Start'].values.reshape(-1, 1))

# Display the DataFrame with the new trend columns
display(workout_summary_df[['DateTime', 'Days_Since_Start',
                            'Smoothed_Distance_meters', 'Distance_Trend',
                            'Smoothed_MovingTime_seconds', 'MovingTime_Trend',
                            'Smoothed_Pace_seconds_per_100m', 'Pace_Trend']].head())
```
